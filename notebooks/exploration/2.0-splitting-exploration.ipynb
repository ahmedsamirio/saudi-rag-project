```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

from src.ingestion import load_pdf

import os
```

```python
RAW_DOCS_PATH = "../../data/raw"
```

```python
docs = [load_pdf(os.path.join(RAW_DOCS_PATH, f)) for f in os.listdir(RAW_DOCS_PATH)]
```

```python
def show_splitter_result(doc, splitter):
    for i, d in enumerate(splitter.split_documents([doc])):
        print('-'*100)
        print('SPLIT', i+1)
        print(d.page_content)
```

```python
splitter = RecursiveCharacterTextSplitter(
    # chunk_size=1000,
    # chunk_overlap=0,
    # separators=['\n\n', '\n', r'\.\s+', ' ', '']
)

show_splitter_result(docs[0], splitter)
```
